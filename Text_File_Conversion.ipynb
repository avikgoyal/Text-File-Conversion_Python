Compare the performance (time taken) of C, C++, Java, R, and Python programs for
1. Convert 200MB, 400 MB, 600 MB, 800 MB, and 1000MB text files to upper case.

```python
#text conversion using python

import os
import datetime
import matplotlib.pyplot as plt
import numpy as np
#creating a file
fp=open('textconversion.txt','w')
q=input("Enter a sentence")
l=len(q)
for i in range (0,l):
  fp.write(str(q[i]))
fp.close()

X=[]
Y=[]
#getting current file size
file_size = os.path.getsize('textconversion.txt')
print("File Size is :", file_size, "bytes")

#converting it into 200MB, 400 MB, 600 MB, 800 MB, and 1000MB

#200mb
with open('textconversion.txt','r') as a,open('newfile.txt','w') as b:
  content=a.read()
  b.write(content * ((10**5)*200))
print ("Writing done !! \nOpen textconversion.txt to view the content")
file_size = os.path.getsize('newfile.txt')
print("File Size is :", file_size/(10**6), "MB")
b.close()
X.append(file_size)
start=datetime.datetime.now()
with open('newfile.txt', 'r') as data_file:
    with open('output.txt', 'a') as output_file:
        for line in data_file:
            output_file.write(line.upper())
end=datetime.datetime.now()
e=end-start
print("The time taken for excution is",e.seconds)
Y.append(e.seconds)

#400mb
with open('textconversion.txt','r') as a,open('newfile.txt','w') as b:
  content=a.read()
  b.write(content * ((10**5)*400))
print ("Writing done !! \nOpen textconversion.txt to view the content")
file_size = os.path.getsize('newfile.txt')
print("File Size is :", file_size/(10**6), "MB")
X.append(file_size)
start=datetime.datetime.now()
with open('newfile.txt', 'r') as data_file:
    with open('output.txt', 'a') as output_file:
        for line in data_file:
            output_file.write(line.upper())
end=datetime.datetime.now()
e=end-start
print("The time taken for excution is",e.seconds)
Y.append(e.seconds)

#600mb
with open('textconversion.txt','r') as a,open('newfile.txt','w') as b:
  content=a.read()
  b.write(content * ((10**5)*600))
print ("Writing done !! \nOpen textconversion.txt to view the content")
file_size = os.path.getsize('newfile.txt')
print("File Size is :", file_size/(10**6), "MB")
X.append(file_size)
start=datetime.datetime.now()
with open('newfile.txt', 'r') as data_file:
    with open('output.txt', 'a') as output_file:
        for line in data_file:
            output_file.write(line.upper())
end=datetime.datetime.now()
e=end-start
print("The time taken for excution is",e.seconds)
Y.append(e.seconds)

#800mb
with open('textconversion.txt','r') as a,open('newfile.txt','w') as b:
  content=a.read()
  b.write(content * ((10**5)*800))
print ("Writing done !! \nOpen textconversion.txt to view the content")
file_size = os.path.getsize('newfile.txt')
print("File Size is :", file_size/(10**6), "MB")
X.append(file_size)
start=datetime.datetime.now()
with open('newfile.txt', 'r') as data_file:
    with open('output.txt', 'a') as output_file:
        for line in data_file:
            output_file.write(line.upper())
end=datetime.datetime.now()
e=end-start
print("The time taken for excution is",e.seconds)
Y.append(e.seconds)

#1000mb
with open('textconversion.txt','r') as a,open('newfile.txt','w') as b:
  content=a.read()
  b.write(content * ((10**5)*1000))
print ("Writing done !! \nOpen textconversion.txt to view the content")
file_size = os.path.getsize('newfile.txt')
print("File Size is :", file_size/(10**6), "MB")
X.append(file_size)
start=datetime.datetime.now()
with open('newfile.txt', 'r') as data_file:
    with open('output.txt', 'a') as output_file:
        for line in data_file:
            output_file.write(line.upper())
end=datetime.datetime.now()
e=end-start
print("The time taken for excution is",e.seconds)
Y.append(e.seconds)

print(X)
print(Y)
plt.plot(X,Y,'o',color="blue")
plt.plot(X,Y)

plt.xlabel('File Size in MB')
plt.ylabel('Time of Execution in Seconds')
plt.title('Text File Conversion For Python')
plt.show()
```
